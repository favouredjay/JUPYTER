```python
import tensorflow as tf
import numpy as np
import logging

logger = tf.get_logger()
logger.setLevel(logging.ERROR)
```

```python
celsius_vals    = np.array([-40, -10,  0,  8, 15, 22,  38],  dtype=float)
fahrenheit_vals = np.array([-40,  14, 32, 46, 59, 72, 100],  dtype=float)

for index, value in enumerate(celsius_vals):
 print ("{} celsius is equal to {} farenheit".format(value, fahrenheit_vals[index]))
```

```python
layer_0 = tf.keras.layers.Dense(units = 1, input_shape = [1])
model = tf.keras.Sequential([layer_0])
model.compile(loss = 'mean_squared_error', optimizer=tf.keras.optimizers.Adam(0.1))
```

```python
trained_model = model.fit(celsius_vals, fahrenheit_vals, epochs = 500, verbose = True)
print("Finished training model")
```

```python
trained_model_1000 = model.fit(celsius_vals, fahrenheit_vals, epochs = 500, verbose = True)
print("Finished training model")
```

```python
import matplotlib.pyplot as plt
plt.xlabel('Epoch Number')
plt.ylabel("Loss Magnitude")
plt.plot(trained_model.history['loss'])
```
